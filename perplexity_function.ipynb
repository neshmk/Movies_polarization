```python
import pandas as pd

df = pd.read_csv("random_movies_1000.csv")
first_10 = df.head(10)

system_prompt = "You are a movie metadata analyst that provides structured info based on movie title and release date."

for index, row in first_10.iterrows():
    title = row['title']
    release_date = row['release_date']

    prompt = f"""
Please analyze the movie below and extract:
- Genre(s)
- Whether the cast or main characters are from marginalized groups (ethnic minorities, women, LGBTQ+)
- A short explanation
- Director 
- Release year
- Title 

Title: {title} ({release_date})

Return your answer in this format:
Title: {title}
Genres: [...]
Marginalized Representation: Yes/No
Details: ...
"""

    result, _ = perplexity(prompt, system_prompt)
    print(f"--- Movie {index + 1} ---")
    print(result)
    print("\n" + "=" * 80 + "\n")
```
